```python
import pandas as pd
import re

# Load the CSV file
input_file = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/Batman (1989)-002.csv"  # Change this to your actual file path
output_file = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/annotated/Annotated-formatted/corrected_file.csv"

# Read the file as a raw text
with open(input_file, "r") as file:
    lines = file.readlines()

# Process each line to ensure correct comma placement
corrected_lines = []
for line in lines:
    # Match the last time value and insert a comma before the text
    corrected_line = re.sub(r"(\d{2}:\d{2}:\d{2}\.\d{3})\s+(\w+)", r"\1, \2", line)
    corrected_lines.append(corrected_line)

# Write the corrected content to a new CSV file
with open(output_file, "w") as file:
    file.writelines(corrected_lines)

print(f"Corrected file saved as {output_file}")
```

Merging the files together

```python
import pandas as pd
import json

json_file =  "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/BATMAN-2/batman.JSON"
csv_file = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/annotated/Annotated-formatted/corrected_file.csv"
output_csv = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/annotated/Merged/data.csv"

with open(json_file, "r") as file:
    json_data = json.load(file)
json_df = pd.DataFrame(json_data["sequences"])
csv_df = pd.read_csv(csv_file)

merged_df = pd.merge(csv_df, json_df, how="left", left_on="Start", right_on="start_frame")

merged_df.to_csv(output_csv, index=False)
print(f"Merged data saved as {output_csv}")
```

associating the keyshots with respect to the sequesces 


```python
import pandas as pd
import os
from natsort import natsorted  # Natural sorting for numerical frame order

# Define paths
csv_file = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/annotated/Annotated-formatted/corrected_file.csv"
keyshots_dir = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/BATMAN-2/keyshots"
output_csv = "/Users/mario/Desktop/Desktop/UofA/4.Winter-2025/ece-910/Batman/annotated/Annotated-formatted/updated_sequences.csv"

# Load the CSV file
df = pd.read_csv(csv_file)

# Create new column for sequences
df["Sequences"] = ""

# Iterate through sequences and map keyshots
all_keyshot_columns = []  # Track all keyshot column names

for i, row in df.iterrows():
    sequence_name = f"sequence_{i+1}"
    sequence_path = os.path.join(keyshots_dir, sequence_name)

    if os.path.exists(sequence_path) and os.path.isdir(sequence_path):
        keyshots = natsorted([f for f in os.listdir(sequence_path) if f.endswith('.jpg')])  # Sort numerically

        # Add sequence name
        df.at[i, "Sequences"] = sequence_name

        # Expand dataframe dynamically with keyshot columns
        for idx, keyshot in enumerate(keyshots):
            col_name = f"Keyshot_{idx+1}"
            if col_name not in df.columns:
                df[col_name] = ""  # Add new columns dynamically
                all_keyshot_columns.append(col_name)

            df.at[i, col_name] = keyshot

# Save updated CSV
df.to_csv(output_csv, index=False)

print(f"✅ Updated CSV saved at: {output_csv}")
```
